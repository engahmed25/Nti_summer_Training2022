```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def costfun (yp , y_points):
    error = yp - y_points
    serror = np.power(error ,2 )
    return np.sum (serror)/(2*len(y_points))
```

```python
def descent (y_points,x_points):
    t0=0
    t1=0
    lt =0.001
    nt=5500
    t0_updata=[0]
    t1_updata=[0]
    cost_updata=[0]
    n=0
    while(n < nt) :
        yp = t1 * x_points +t0
        cost = costfun( yp , y_points )
        g0=(np.sum(yp -y_points)/ len(y_points))
        g1=(np.sum((yp -y_points)*x_points) / len(y_points))
        t0 = t0 - lt*g0
        t1 = t1 - lt *g1
        t0_updata.append(t0)
        t1_updata.append(t1)
        cost_updata.append(cost)
        n+=1
        if (round(cost_updata[-1],5)==round(cost_updata[-2],5)):
            break
    print( "x ={}  y ={} z={} ".format(t0 , t1 ,cost))  
    print("\n",n)
    plt.clf()
    fig , ax = plt.subplots(1,2)
    ax[0].plot(t0_updata  , cost_updata)
    ax[0].set(xlabel='t0', ylabel='cost')
    ax[1].plot(t1_updata , cost_updata)
    ax[1].set(xlabel='t1', ylabel='cost');
    plt.show()
    plt.clf()
    plt.plot( cost_updata)
    plt.xlabel("Iterations")
    plt.ylabel("Cost")
    plt.title("How the cost decreases")
    plt.show()
```

```python
x_points = np.array([[1,4,2,3],[1,5,8,5]])
y_points = np.array([1,2,3,1])
```
